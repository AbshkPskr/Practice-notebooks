```
ApiKeyOmbd = "b01fbe14"
client = OMDBClient(apikey = ApiKeyOmbd)
```

```
MovieListOmdb =client.search("star wars")
```

```
MovieListOmdbComplete = []
for  mov in MovieListOmdb:
  MovieListOmdbComplete.append(client.get(imdbid=mov["imdb_id"]))
```

```
data = pd.DataFrame(MovieListOmdbComplete)
```
