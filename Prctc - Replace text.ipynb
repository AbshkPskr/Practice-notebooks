```
#input string
Input1 = "asdfasf bcz asdfasdf bcz bcz"
```

```
#using replace method
Output1 = Input1.replace('bcz','because')
Output1
```

```
#replacing in list of strings using for loop
Input2 = ["asdfasf bcz asdfasdf bcz bcz",
          "asdfasf bcz asdfasdf bcz bcz",
          "asdfasf bcz asdfasdf bcz bcz",
          "asdfasf bcz asdfasdf bcz bcz"]

Output2 = []

for i in  Input2:
  Output2.append(i.replace('bcz','because'))
  
Output2
```

```
#replacing using different set of words in dictionary

Dict1 = {'bcz':'because','tk':'take'}
Input3 = "asdfasf bcz tk asdfasdf bcz bcz tk"
Output3 = ''

for i,j in Dict1.items():
  Input3 = Input3.replace(i,j)
  
  
Output3 = Input3
Output3
```

```
#replacing in Dataframe
import pandas as pd

#creating dataframe
DF = pd.DataFrame({'Words':["asdf bcz bcz","bcz","bcz bcz"]},index = [0,1,2])
```

```
#Looping through all the values in dataframe.

for i in range(0,len(DF['Words'])):
  DF['Words'][i] = DF['Words'][i].replace('bcz','because')
```
