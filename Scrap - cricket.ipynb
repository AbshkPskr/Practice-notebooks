```
import urllib3
from bs4 import BeautifulSoup
```

```
import requests
```

```
url = "https://stats.espncricinfo.com/ci/engine/records/batting/most_runs_career.html?id=1;type=trophy"
webpage = requests.get(url)
webpage.text
```

```
soup = BeautifulSoup(webpage.text, 'html.parser')
print(soup.prettify())
```

```
myTable = soup.find('table',{'class':'engineTable'})
```

```
dataTags = myTable.findAll('td')
```

```
player = []
span = []
match = []
innings = []
notouts = []
runs = []
highestinns = []
avg = []
hundreds = []
fifties = []
ducks = []
for i in range(0,len(dataTags),11):
    player.append(dataTags[i].text)
    span.append(dataTags[i+1].text)
    match.append(int(dataTags[i+2].text))
    innings.append(int(dataTags[i+3].text))
    notouts.append(int(dataTags[i+4].text))
    runs.append(int(dataTags[i+5].text))
    highestinns.append(int(dataTags[i+6].text[0:-1]))
    avg.append(float(dataTags[i+7].text))
    hundreds.append(int(dataTags[i+8].text))
    fifties.append(int(dataTags[i+9].text))
    ducks.append(int(dataTags[i+10].text))
```

```
import pandas as pd
data = pd.DataFrame({'player name':player,
              'playing span':span,
              'matches played':match,
              'innings batted':innings,
              'not outs':notouts,
              'runs scored':runs,
              'highest inns score':highestinns,
              'batting average':avg,
              'hundreds scored':hundreds,
              'fifties scored':fifties,
              'ducks scored':ducks})
```
