```
import requests
from bs4 import BeautifulSoup
```

```
url = "https://en.wikipedia.org/wiki/Comparison_of_multi-paradigm_programming_languages"
webpage = requests.get(url)
webpage.text
```

```
soup = BeautifulSoup(webpage.text, 'html.parser')
print(soup.prettify())
```

```
myTable = soup.find('table',{'class':'wikitable sortable'})
```

```
myTable
```

```
links = myTable.findAll('a')
```

```
links
```

```
languages = []
for link in links:
  languages.append(link.get('title'))
```

```
import pandas as pd
df = pd.DataFrame()
```

```
df['languages'] =languages
```

```
df = df.dropna()
df = df.reset_index()
```
